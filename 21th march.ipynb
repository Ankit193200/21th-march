# Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose one over the other.

Ordinal Encoding: Assigning a unique integer to each category based on its ordinal relationship. The order matters.

Label Encoding: Assigning a unique integer to each category without considering any ordinal relationship.

Example:

If you have a variable "Size" with categories "Small," "Medium," and "Large":
Ordinal Encoding: Assign 1, 2, and 3 based on the order (Small: 1, Medium: 2, Large: 3).
Label Encoding: Assign any unique integers to each category without considering the order (e.g., Small: 2, Medium: 1, Large: 3).
Choosing Between:

Choose Ordinal Encoding when there is a clear order or hierarchy among categories.
Choose Label Encoding when categories are nominal and have no inherent order.
# Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.

Target Guided Ordinal Encoding: Assigning ranks to categories based on their mean or median of the target variable.
Example:

In a binary classification project predicting customer churn:
Calculate the mean target value for each category of a categorical variable.
Assign ranks based on the mean values.
Higher ranks for categories with a higher probability of the target.
Use Case:

Use it when there's a significant relationship between the categorical variable and the target, and you want to capture that in the encoding.

# Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium, large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library. Show your code and explain the output.

python
Copy code
from sklearn.preprocessing import LabelEncoder

    # Sample data
data = {'Color': ['red', 'green', 'blue', 'red'],
        'Size': ['small', 'medium', 'large', 'medium'],
        'Material': ['wood', 'metal', 'plastic', 'metal']}

    # Creating a DataFrame
import pandas as pd
df = pd.DataFrame(data)
 
    # Label encoding
label_encoder = LabelEncoder()

for column in df.columns:
    df[column+'_encoded'] = label_encoder.fit_transform(df[column])

    # Output
print(df[['Color', 'Color_encoded', 'Size', 'Size_encoded', 'Material', 'Material_encoded']])
Explanation:

The code uses LabelEncoder to encode categorical variables.
Each unique category is assigned a unique integer.
The output DataFrame shows the original categories and their corresponding encoded values.
